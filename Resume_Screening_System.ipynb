# Resume Screening and Ranking System

This system automatically screens and ranks resumes based on a given job description using NLP and ML techniques.

## Key Features:
- **Text Cleaning & Parsing**: Sanitizes resume and JD text.
- **Skill Extraction**: Uses spaCy Matcher to identify technical skills.
- **Similarity Scoring**: TF-IDF & Cosine Similarity for semantic matching.
- **Candidate Ranking**: Weighted scoring based on role fit.
- **Skill Gap Identification**: Highlights missing skills for each candidate.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from ranker import ResumeRanker
import warnings
warnings.filterplus = 'ignore'
```

### 1. Initialize System and Load Data

```python
ranker = ResumeRanker()

# Sample Dataset (Replace with actual CSV loading if available)
data = {
    'Candidate_Name': ['Alice Johnson', 'Bob Smith', 'Charlie Brown', 'Diana Prince', 'Evan Wright'],
    'Resume_Text': [
        'Data Scientist with 5 years experience in Python, Machine Learning, and SQL. Expert in Deep Learning and NLP.',
        'Senior Java Developer. Extensive knowledge in Spring Boot, SQL, and AWS infrastructure.',
        'Junior Python Developer. Familiar with Web Development (HTML/CSS), Javascript, and basic Data Analysis with Pandas.',
        'Machine Learning Engineer with strong background in PyTorch, TensorFlow, and Azure.',
        'HR Manager with excellence in recruitment, talent acquisition, and Excel.'
    ]
}
df = pd.DataFrame(data)
```

### 2. Enter Job Description

```python
job_description = """
We are looking for a Data Scientist / ML Engineer. 
Required Skills: Python, Machine Learning, SQL, AWS, and Deep Learning.
"""
print(f"Target Job Description: {job_description}")
```

### 3. Rank Candidates

```python
ranked_results = ranker.rank_candidates(job_description, df)

# Display Top Candidates
display(ranked_results[['Candidate_Name', 'Final_Score', 'Matched_Skills', 'Missing_Skills']].head())
```

### 4. Visual Comparison

```python
plt.figure(figsize=(10, 6))
sns.barplot(x='Final_Score', y='Candidate_Name', data=ranked_results, palette='viridis')
plt.title('Candidate Ranking Based on Role Fit')
plt.xlabel('Score (0.0 - 1.0)')
plt.xlim(0, 1)
plt.show()
```

### 5. Skill Gap Analysis
Recruiters can use this to see what specific skills are missing in each candidate.

```python
for index, row in ranked_results.iterrows():
    print(f"--- {row['Candidate_Name']} ---")
    print(f"Match Score: {row['Final_Score']:.2f}")
    print(f"Found Skills: {', '.join(row['Matched_Skills']) if row['Matched_Skills'] else 'None'}")
    print(f"Missing Skills: {', '.join(row['Missing_Skills']) if row['Missing_Skills'] else 'None'}")
    print("\n")
```
